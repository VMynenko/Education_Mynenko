```python
import requests, json
import pandas as pd

baseurl = 'https://tasktrackerapimanager.azure-api.net/tasks/tasks/activity/'

headers = {

    'X-Subscription-Key': '665bf4f4be25470a807c2467e2db8706'

}

pg = 0
req = requests.get(baseurl + f'{pg}', headers=headers)

data = req.json()
```

```python
data
```

```python
pages = data['page']['last']
pages
```

```python
try:
    my_list = []
    while pg < pages:
        req = requests.get(baseurl + f'{pg}', headers=headers)
        data = req.json()
        for i in data['activities']:
            my_list.append(i)
        pg += 1
except Exception as e:
    print(e)
```

```python
while pg <= pages:
    req = requests.get(baseurl + f'{pg}', headers=headers)
    data = req.json()
    for i in data['activities']:
        my_list.append(i)
    pg += 1
    if i == KeyError:
        continue
```

```python
df = pd.DataFrame(my_list)
df
```

```python
df_task = df['task'].apply(pd.Series)
df_task
```

```python
full_df = pd.merge(df, df_task, left_index=True, right_index=True)
full_df
```

```python
del full_df['task']
```

```python
full_df
```

```python
df_done = full_df.query('type_x == "Done"')
```

```python
df_progressStarted = full_df.query('type_x == "ProgressStarted"')
df_progressStarted
```

```python
df_progressStarted = df_progressStarted[['timeStamp', 'type_x', 'id']]
```

```python
new_full_df = df_done.merge(df_progressStarted, on='id')
```
