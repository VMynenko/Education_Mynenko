```python
german_orders = retail.query('Country == "Germany"') \
    .groupby('CustomerID').agg({'InvoiceNo' : pd.Series.nunique})
```

```python
german_orders
```

```python
perc_80 = german_orders.InvoiceNo.quantile(q=0.8)
```

```python
perc_80
```

```python
germany_top = german_orders.query('InvoiceNo > @perc_80').index
```

```python
germany_top
```

```python
 top_retail_germany = retail.query('CustomerID in @germany_top')
```

```python
top_retail_germany
```

```python
top_retail_germany.groupby('StockCode').agg({'InvoiceNo' : 'count'}).sort_values('InvoiceNo', ascending=False)
```

```python
top_retail_germany.StockCode.value_counts()
```
